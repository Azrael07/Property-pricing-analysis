```python
# Importing necessary libraries
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = (5.0,5.0)
```

```python
path = 'boston.csv'
df = pd.read_csv(path)
df.head()
```

```python
df.shape
```

```python
# Convert it into pandas dataframe
data = pd.DataFrame(df)
data.head()
```

```python
corr = data.corr(method = 'pearson')
corr
```

```python
#corr.style.background_gradient(cmap='coolwarm')
corr.style.background_gradient(cmap='coolwarm').set_precision(2)
```

```python
# Statistical summary
data.describe()
```

```python
# Target (dependent variable) and features (independent variables)
Y = data["PRICE"]
X = data.drop("PRICE", axis = 1)
```

```python
#Split the data
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size = 0.3, random_state = 0)
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

```python
#X_train["PRICE"] = y_train
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.fit_transform(X_test)
```

```python
#Linear Regression model
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
model = LinearRegression() 
model.fit(X_train_scaled, y_train)

#To calculate the R squared
R2 = model.score(X_test_scaled, y_test)
print("R-squared: %.4f" % R2) 

print('intercept (b0):', model.intercept_)
print('slope (b1):', model.coef_)

y_pred = model.predict(X_test_scaled) 

#Find the mean square error
MSE = mean_squared_error(y_test, y_pred)
print("Mean squared error: %.2f" % MSE )
```

```python
#Stochastic Gradient Decent(SGD) for Linear Regression
from sklearn.linear_model import SGDRegressor
from sklearn.metrics import mean_squared_error
model2 = SGDRegressor()
model2.fit(X_train_scaled, y_train)

R2 = model2.score(X_test_scaled, y_test)
print("R-squared: %.4f" % R2) 

y_pred2 = model2.predict(X_test_scaled)

MSE = mean_squared_error(y_test, y_pred2)
print("Mean squared error: %.2f" % MSE )

print("Coefficients: \n", model2.coef_)
print("Y_intercept", model2.intercept_)
```

```python
#Stochastic Gradient Decent(SGD) for Linear Regression
from sklearn.linear_model import SGDRegressor
from sklearn.metrics import mean_squared_error
model5 = SGDRegressor(alpha=0.0001 , loss= 'squared_loss', n_iter=20000, penalty = 'l2')
model5.fit(X_train_scaled, y_train)

R2 = model5.score(X_test_scaled, y_test)
print("R-squared: %.4f" % R2) 

y_pred2 = model5.predict(X_test_scaled)

MSE = mean_squared_error(y_test, y_pred2)
print("Mean squared error: %.2f" % MSE )

print("Coefficients: \n", model5.coef_)
print("Y_intercept", model5.intercept_)
```
